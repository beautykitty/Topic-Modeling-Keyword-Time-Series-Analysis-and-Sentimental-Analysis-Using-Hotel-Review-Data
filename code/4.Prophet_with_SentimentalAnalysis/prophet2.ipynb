```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/공모전/tmp/processed_data_2_topic_5_1.csv', encoding='euc-kr')
```

```python
df
```

```python
df['stay_date'] = pd.to_datetime(df['stay_date'])
df.sort_values(by='stay_date', inplace=True)
df.reset_index(inplace=True)
```

```python
df
```

```python
df_reg = df[['stay_date','sentimental_score_negative']].rename(columns = {'stay_date':'ds', 'sentimental_score_negative': 'y'})
```

```python
from fbprophet import Prophet
model1 = Prophet(changepoint_prior_scale=10, 
                changepoints=['2020-02-01'], 
                yearly_seasonality=10, 
                weekly_seasonality=False, 
                daily_seasonality=False, 
                growth='linear')
model1.add_country_holidays(country_name='KR')
```

```python
df_reg = df_reg.pivot_table(index='ds', aggfunc='median').reset_index()
```

```python
future = model1.fit(df_reg).make_future_dataframe(periods=0, freq='M')
forecast = model1.predict(future)
fig1 = model1.plot(forecast)
fig2 = model1.plot_components(forecast)
```
