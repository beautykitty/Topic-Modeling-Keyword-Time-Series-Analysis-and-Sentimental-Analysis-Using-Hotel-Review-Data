```python
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/공모전/tmp/sentimental_per_topic.csv', encoding='euc-kr')
```

```python
data.head()
```

```python
del data['Unnamed: 0']
```

```python
data.sentimental_score[0]
```

```python
import ast

data['sentimental_score'] = data.sentimental_score.apply(lambda x: ast.literal_eval(x))
```

```python
data['sentimental_score_negative'] = data['sentimental_score'].apply(lambda x: x[0][0])
```

```python
data['sentimental_score_positive'] = data['sentimental_score'].apply(lambda x: x[0][1])
```

```python
data.topic_score
```

```python
data['topic_score'] = data['topic_score'].apply(lambda x: x[1:-1].split())
```

```python
data['topic_score']
```

```python
for i in range(10):
    data[f'topic_score_#{i}'] = data['topic_score'].apply(lambda x: float(x[i]))
```

```python
data.info()
```

```python
import ast

data['sentimental_per_topic'] = data.sentimental_per_topic.apply(lambda x: ast.literal_eval(x))
```

```python
data['sentimental_per_topic'][0][0]
```

```python
for i in range(10):
    data[f'topic_score_#{i}_negative'] = data['sentimental_per_topic'].apply(lambda x: x[i][0])
```

```python
for i in range(10):
    data[f'topic_score_#{i}_positive'] = data['sentimental_per_topic'].apply(lambda x: x[i][1])
```

```python
data2 = data.drop(['sentimental_score', 'sentimental_per_topic', 'topic_score'], axis=1)
```

```python
data2.info()
```

```python
data2.to_csv('/content/drive/MyDrive/공모전/tmp/processed_data_2.csv', encoding='euc-kr')
```
